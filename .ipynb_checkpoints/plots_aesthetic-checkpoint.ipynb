```python
import matplotlib.pyplot as plt
import numpy as np

# Configuración global de Matplotlib
plt.rcParams['figure.figsize'] = (8, 6)  # Tamaño predeterminado de la figura (ancho x alto)
plt.rcParams['font.size'] = 12           # Tamaño de fuente predeterminado
plt.rcParams['axes.labelsize'] = 12      # Tamaño de fuente de etiquetas de ejes
plt.rcParams['axes.titlesize'] = 14      # Tamaño de fuente del título del gráfico
plt.rcParams['xtick.labelsize'] = 10     # Tamaño de fuente de etiquetas del eje x
plt.rcParams['ytick.labelsize'] = 10     # Tamaño de fuente de etiquetas del eje y

# Crear datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Crear el gráfico
plt.plot(x, y, label='Sen(x)', linewidth=2, color='b', linestyle='--', marker='o', markersize=6, markerfacecolor='r')

# Etiquetas de ejes y título
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de Sen(x)')

# Leyenda
plt.legend()

# Rejilla
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Configuración de colores pasteles y primarios
colors = ['#FF6F61', '#6B4226', '#A9A9A9']  # Colores pasteles
highlight_color = '#1E90FF'  # Color primario para resaltar

# Configuración global de Matplotlib
plt.rcParams['figure.figsize'] = (8, 6)
plt.rcParams['font.size'] = 12
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 10

# Crear datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Crear el gráfico con colores y transparencias
plt.plot(x, y, label='Sen(x)', linewidth=2, color=colors[0], linestyle='--', marker='o', markersize=6, markerfacecolor=highlight_color, alpha=0.7)

# Etiquetas de ejes y título
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de Sen(x)')

# Leyenda
plt.legend()

# Rejilla interna
plt.grid(True, which='both', linestyle='--', linewidth=0.5, color='gray', alpha=0.5)

# Configuración de fondo
plt.gca().set_facecolor('#F5F5F5')  # Color de fondo estilo pastel

# Mostrar el gráfico
plt.show()
```
