# Correlaciones con Oulu

```python
##### FILTRADOOOOO
# Filtrando los datos para el rango de aos DEL DETECTOR
start_date = '1990-01-01'
end_date = '2021-09-30'
oulu_corr = merged[(merged['date'] >= start_date) & (merged['date'] <= end_date)]
oulu_corr.set_index('date', inplace=True)
oulu_corr = oulu_corr.astype('float64') # 
```

```python
configure_plots()
# 1. Calcular los valores medios de ambas variables
mean_counts_o = oulu_corr["oulu"].mean()
mean_sunspots_o = oulu_corr["sunspots"].mean()
# 2. Calcular los desplazamientos (\Delta)
max_lag = 30  # Establece el nmero mximo de desplazamientos a considerar (30 das para incluir hasta el ciclo de hasta ~27 das)
# 3. Calcular las desviaciones estndar normalizadas (\delta_{p} y \delta_{q})
std_counts_o = oulu_corr["oulu"].std()
std_sunspots_o = oulu_corr["sunspots"].std()
# 4. Inicializar una lista para almacenar los coeficientes de correlacin cruzada
cross_correlations_o = []
# 6. Aplicar la frmula de correlacin cruzada para cada desplazamiento (\Delta)
for lag in range(-max_lag, max_lag + 1):
    sum_product_o = 0
    #
    for i in range(len(oulu_corr) - abs(lag)):
        diff_counts_o = oulu_corr["oulu"].iloc[i] - mean_counts_o
        diff_sunspots_o = oulu_corr["sunspots"].iloc[i + lag] - mean_sunspots_o
        sum_product_o += diff_counts_o * diff_sunspots_o
    # Clculo del coeficiente de correlacin cruzada para el desplazamiento actual
    cross_corr_o = (1 / (len(oulu_corr) - 1)) * (sum_product_o / (std_counts_o * std_sunspots_o))
    cross_correlations_o.append(cross_corr_o)

# Ahora cross_correlations contiene los coeficientes de correlacin cruzada para diferentes desplazamientos
# Imprimir valor para un desplazamiento de 0:
lag_0_corr_o = cross_correlations_o[max_lag]
print("Coeficiente de correlacin cruzada para desplazamiento 0:", lag_0_corr_o)
```

```python
lag_range_o = list(range(-max_lag, max_lag + 1))
plt.plot(lag_range_o, cross_correlations_o, marker='o', linestyle='-', color='darkslateblue')  # Personaliza el color a tu eleccin
plt.xlabel('Lag (das)')
plt.ylabel('Coeficiente de Correlacin')
plt.title('CORRELACIN DEL ARTCULO')
plt.grid(True, linestyle='--', alpha=0.7)  # Aade una grilla con estilo punteado y transparencia
plt.tight_layout()  # Ajusta el espacio para que todos los elementos se muestren correctamente
plt.show()
```

## Correlacin con un lag mensual

```python
lags = np.arange(30, 30*12*25, 30)  # Lags de 1 mes a 12 aos (en das)
correlations = []

for lag in lags:
    # Desplazar una de las columnas para calcular la correlacin con lag
    shifted = oulu_corr['sunspots'].shift(lag)
    correlation = oulu_corr['oulu'].corr(shifted)
    correlations.append(correlation)
# Convertir lags a aos para la grfica
lags_years = lags / 365
# Crear un grfico de los coeficientes de correlacin
plt.figure(figsize=(10, 5))
plt.plot(lags_years, correlations)
plt.xlabel('Lag (aos)')
plt.ylabel('Coeficiente de correlacin de Pearson (mensual)')
plt.title('Correlacin entre CRI y sunspots')
plt.show()
```

```python
import numpy as np
import pandas as pd

# Parmetros
amplitude1 = 1.0
amplitude2 = 1.0
frequency = 0.1  # Frecuencia en ciclos por unidad de tiempo
phase_shift = np.pi  # Desfase de 45 grados para generar anticorrelacin
num_points = 100

# Generar el tiempo
time = np.arange(0, num_points)

# Generar las dos series sinusoidales
series1 = amplitude1 * np.sin(2 * np.pi * frequency * time)
series2 = amplitude2 * np.sin(2 * np.pi * frequency * time + phase_shift)

# Crear un DataFrame
data = {'Time': time, 'Series1': series1, 'Series2': series2}
test = pd.DataFrame(data)

# Aadir ruido aleatorio para hacerlo ms realista (opcional)
#noise = np.random.normal(0, 0.1, num_points)
#test['Series1'] += noise
#test['Series2'] += noise

# Visualizar las primeras filas del DataFrame
print(test.head())

# Ahora puedes usar el cdigo de correlacin cruzada que mencionaste para analizar estas series.
```

## 1) Correlacin a mano

```python
# Establece el rango mximo de desplazamiento que deseas explorar
max_lag = 5  # Puedes ajustar esto segn tus necesidades

# Inicializa una lista para almacenar los coeficientes de correlacin cruzada
cross_correlations_t = []


# Calcula las medias de ambas series
mean_CRI_t = test['Series1'].mean()
mean_sunspots_t = test['Series2'].mean()

# Calcula las desviaciones estndar de ambas series
std_CRI_t = test['Series1'].std()
std_sunspots_t = test['Series2'].std()

# 6. Aplicar la frmula de correlacin cruzada para cada desplazamiento (\Delta)
for lag in range(-max_lag, max_lag + 1):
    sum_product_t = 0
    valid_data_count_t = 0

    for i in range(len(test) - abs(lag)):
        diff_CRI_t = test['Series1'].iloc[i] - mean_CRI_t
        diff_sunspots_t = test['Series2'].iloc[i + lag] - mean_sunspots_t

        if not (np.isnan(diff_CRI_t) or np.isnan(diff_sunspots_t)):
            sum_product_t += diff_CRI_t * diff_sunspots_t
            valid_data_count_t += 1

    # Calcular el coeficiente de correlacin cruzada para el desplazamiento actual
    if valid_data_count_t > 0:
        cross_corr_t = (1 / (valid_data_count_t - 1)) * (sum_product_t / (std_CRI_t * std_sunspots_t))
        cross_correlations_t.append(cross_corr_t)
    else:
        cross_correlations_t.append(np.nan)
```

```python
lag_range_t = list(range(-max_lag, max_lag + 1))
plt.plot(lag_range_t, cross_correlations_t, marker='o', linestyle='-', color='darkslateblue')  # Personaliza el color a tu eleccin
plt.xlabel('Lag (tiempo)')
plt.ylabel('Coeficiente de Correlacin')
plt.title('TEST Correlacin Cruzada')
plt.grid(True, linestyle='--', alpha=0.7)  # Aade una grilla con estilo punteado y transparencia
plt.tight_layout()  # Ajusta el espacio para que todos los elementos se muestren correctamente
plt.show()
```

# 2) Correlacin hecha con Scipy

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Funcin para generar fechas desde una fecha inicial hasta una fecha final con un intervalo de das
def generate_dates(start_date, end_date, delta):
    current_date = start_date
    while current_date <= end_date:
        yield current_date
        current_date += delta

# Definir las fechas de inicio y fin
start_date = datetime(2006, 1, 1)
end_date = datetime(2021, 11, 30)

# Generar fechas con un intervalo diario
dates = list(generate_dates(start_date, end_date, timedelta(days=1)))

# Crear el DataFrame para la seal sinusoidal con periodicidad de 11 aos
df_sinusoidal = pd.DataFrame({
    'date': dates,
    'CRI': np.sin(2 * np.pi * np.arange(len(dates)) / (11 * 365.25))  # Seal sinusoidal con periodicidad de 11 aos
})

# Crear el DataFrame para la seal sinusoidal completamente anticorrelacionada
df_anticorrelated = pd.DataFrame({
    'date': dates,
    'CRI': -np.sin(2 * np.pi * np.arange(len(dates)) / (11 * 365.25))  # Seal sinusoidal anticorrelacionada
})

# Mostrar las primeras filas de ambos DataFrames
print("DataFrame con seal sinusoidal:")
print(df_sinusoidal.head())

print("\nDataFrame con seal sinusoidal anticorrelacionada:")
print(df_anticorrelated.head())
```

```python
plt.plot(df_sinusoidal['CRI'])
plt.plot(df_anticorrelated['CRI'])
```

```python
lags = np.arange(-30*12*12, 30*12*12, 30)  # Lags de -12 aos a 12 aos (en das)
correlations = []

for lag in lags:
    # Desplazar una de las columnas para calcular la correlacin con lag
    shifted = df_sinusoidal['CRI'].shift(lag)
    correlation = df_anticorrelated['CRI'].corr(shifted)
    correlations.append(correlation)

# Convertir lags a aos para la grfica
lags_years = lags / 365.25

# Crear un grfico de los coeficientes de correlacin
plt.figure(figsize=(8, 6))  # Ajust el tamao del grfico para una mejor visualizacin
plt.plot(lags_years, correlations, marker='.', linestyle='-', color='darkslateblue')
plt.xlabel('Lag (aos)')
plt.ylabel('Coeficiente de Correlacin')
plt.ylim(-1, 1)
plt.title('')

# Aadir lneas verticales en 5.5 y 11 aos
plt.axvline(x=5.5, color='lightcoral', linestyle='--', label='')
plt.axvline(x=11, color='lightcoral', linestyle='--', label='')
plt.axvline(x=-5.5, color='lightcoral', linestyle='--', label='')
plt.axvline(x=-11, color='lightcoral', linestyle='--', label='')
plt.axvline(x=1, color='lightcoral', linestyle='--', label='')


# Aadir etiquetas para los aos en las lneas verticales
plt.text(5.5, 0.2, '5.5 aos', rotation=90, color='lightcoral', fontsize=10, ha='right', va='bottom')
plt.text(11, 0.2, '11 aos', rotation=90, color='lightcoral', fontsize=10, ha='right', va='bottom')
plt.text(-5.5, 0.2, '5.5 aos', rotation=90, color='lightcoral', fontsize=10, ha='right', va='bottom')
plt.text(-11, 0.2, '11 aos', rotation=90, color='lightcoral', fontsize=10, ha='right', va='bottom')

plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.legend()  # Aadir leyenda
#plt.savefig('corr_auger_sn.pdf')

plt.show()
```
