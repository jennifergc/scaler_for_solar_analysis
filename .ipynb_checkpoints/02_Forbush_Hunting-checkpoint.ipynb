```python
eventoFDU = scaler3600s[scaler3600s['Date'] >= "2011-02-17"]
evento180211 = eventoFDU[eventoFDU['Date'] <= "2011-02-21" ]
evento180211.plot(x='Date',y='ScalerCorr',color='indigo',rot=30,figsize=(10, 8), linewidth =2, label ="Auger")

# Configura los valores de fecha que deseas mostrar en el eje x
dates_to_show = ["2011-02-17", "2011-02-19", "2011-02-21"]
# Configura el formato de las etiquetas del eje x
plt.xticks(dates_to_show, rotation=0)

plt.xlabel('Date', fontsize=35)  # Nombre del eje horizontal
plt.ylabel('Cosmic Ray Intensity', fontsize=35)  # Nombre del eje vertical


plt.savefig('scaler_180211.png')
plt.show()
```
