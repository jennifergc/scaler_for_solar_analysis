```python
import numpy as np
import matplotlib.pyplot as plt

def scalerfft_period(sample, dataframe, marks, name, scalercol, conf_interval=False):
    data = dataframe[scalercol]
    fft_result = np.fft.fft(data)
    power_spectrum = np.abs(fft_result) ** 2
    sampling_rate = sample
    n = len(data)
    frequencies = np.fft.fftfreq(n, d=sampling_rate)

    periods = 1 / frequencies
    periods_days = periods / 86400

    periods_to_mark = []  # Define la variable incluso si no se marca ninguna frecuencia

    if conf_interval:
        # Bootstrap para calcular el intervalo de confianza
        num_bootstrap_samples = 100
        bootstrap_power_spectra = np.zeros((num_bootstrap_samples, n))

        for i in range(num_bootstrap_samples):
            bootstrap_data = np.random.choice(data, n, replace=True)
            bootstrap_fft = np.fft.fft(bootstrap_data)
            bootstrap_power_spectra[i] = np.abs(bootstrap_fft) ** 2

        # Calcular el percentil 2.5 y 97.5 para el intervalo de confianza
        lower_percentile = np.percentile(bootstrap_power_spectra, 2.5, axis=0)
        upper_percentile = np.percentile(bootstrap_power_spectra, 97.5, axis=0)

    periods_days_sorted = sorted(periods_days, reverse=True)
    power_spectrum_sorted = [power_spectrum[i] for i in np.argsort(periods_days)[::-1]]

    plt.figure(figsize=(10, 6))
    plt.plot(periods_days_sorted, power_spectrum_sorted, label='Espectro de Potencias')

    if conf_interval:
        plt.fill_between(periods_days_sorted, lower_percentile, upper_percentile, color='gray', alpha=0.5, label='Intervalo de Confianza')

    plt.yscale('log')
    plt.xscale('log')
    plt.gca().invert_xaxis()
    plt.xlabel('Período (días)')
    plt.ylabel('Potencia')
    plt.grid()

    if marks:
        for period in periods_to_mark:
            plt.axvline(x=period, color='red', linestyle='--', label=f'Frecuencia {period} nHz')
            plt.text(period, 1, f'{period} d', rotation=50, va='bottom', ha='center')

    plt.legend()
    archivo_salida = f'{name}.png'
    plt.savefig(archivo_salida)
    plt.show()
```

```python
# Llama a la función con conf_interval=True para calcular el intervalo de confianza
scalerfft_period(300,scaler300s, 1, 'prueba_fft', 'ScalerCorr', conf_interval=True)
```

```python
import numpy as np
from scipy import signal
from scipy.fft import fft, fftshift
import matplotlib.pyplot as plt

# Asume que 'base' es tu DataFrame de pandas y 'counts/s' es tu columna de interés
time_series = base['counts/s']

# Define una condición para los valores que quieres ignorar
# Por ejemplo, puedes ignorar los valores que son NaN (Not a Number)
condition = np.isnan(time_series)
#condition = (time_series == 0)

# Ignora los valores que cumplen la condición
time_series = time_series[~condition]

# Calcula la función de autocorrelación de la serie de tiempo
autocorr = np.correlate(time_series, time_series, mode='full')

# Aplica la ventana de Hanning a la función de autocorrelación
window = signal.windows.hann(len(autocorr))
autocorr_windowed = autocorr * window

# Calcula la transformada de Fourier de la función de autocorrelación con ventana
spectrum = np.abs(fft(autocorr_windowed))


# Calcula las frecuencias correspondientes a cada punto del espectro
frequencies = np.fft.fftfreq(len(spectrum), d=1.0)  # Asume que el muestreo es de 1 hora

# Convierte las frecuencias a periodos en días
periods = 1 / (frequencies * 24)  # Convierte las horas a días


# Grafica el espectro de potencia
plt.plot(periods, fftshift(spectrum))
plt.title("Espectro de potencia")
plt.xscale('log')  # Utilizar escala logarítmica en el eje x
plt.yscale('log')
plt.show()

```

```python
import numpy as np
from scipy.fft import fft, fftshift

def blackman_tukey_hanning(df, M=None):
    # Paso 1: Calcula la función de autocorrelación
    autocorr = np.correlate(df, df, mode='full')
    
    # Paso 2: Aplica la ventana de Hanning
    if M is None:
        M = len(df) // 2
    window = np.hanning(M)
    autocorr_windowed = autocorr[:M] * window
    
    # Paso 3: Calcula la Transformada de Fourier
    psd = np.abs(fft(autocorr_windowed))
    
    return psd

```

```python

# Aplica el método de Blackman-Tukey con ventana de Hanning a tu serie de tiempo
psd = blackman_tukey_hanning(base['counts/s'])
```
