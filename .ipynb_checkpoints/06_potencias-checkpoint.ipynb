### ARIMA 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller

# Comprobar si los datos son estacionarios
result = adfuller(data)
print(f'ADF Statistic: {result[0]}')
print(f'p-value: {result[1]}')

# Si los datos no son estacionarios, realiza la diferenciación
if result[1] > 0.05:
    data_diff = data.diff().dropna()  # Realiza la diferenciación
else:
    data_diff = data  # Los datos ya son estacionarios

# Graficar la serie temporal y su autocorrelación y autocorrelación parcial
plt.figure(figsize=(12, 4))
plt.subplot(131)
plt.plot(data)
plt.title('Serie Temporal Original')
plt.subplot(132)
plot_acf(data, lags=30, ax=plt.gca())
plt.title('Autocorrelación')
plt.subplot(133)
plot_pacf(data, lags=30, ax=plt.gca())
plt.title('Autocorrelación Parcial')
plt.tight_layout()
plt.show()

# Ajustar un modelo ARIMA a los datos diferenciados
p, d, q = 1, 1, 1  # Valores de p, d y q para ARIMA
model = ARIMA(data_diff, order=(p, d, q))
results = model.fit()

# Imprimir el resumen del modelo
print(results.summary())

# Predecir valores futuros con el modelo
forecast_steps = 24  # Por ejemplo, para predecir 24 pasos hacia adelante
forecast = results.forecast(steps=forecast_steps)

# Graficar las predicciones
plt.figure(figsize=(8, 4))
plt.plot(data_diff, label='Datos Diferenciados')
plt.plot(range(len(data_diff), len(data_diff) + forecast_steps), forecast, label='Predicciones', color='red')
plt.title('Predicciones con Modelo ARIMA')
plt.legend()
plt.show()
```

```python
import numpy as np
import pywt
import pandas as pd

# Suponiendo que 'scaler3600' es tu DataFrame y 'ScalerCorr' es la columna de interés
# Suponemos que ya tienes los datos en un DataFrame llamado 'scaler3600'
data = scaler3600s['ScalerCorr'].values  # Convertir la columna en un arreglo de numpy

# Realizar una Transformada de Wavelet Discreta (DWT)
wavelet = 'haar'  # Tipo de wavelet (puedes elegir otro)
level = 5  # Nivel de descomposición de la DWT

# Aplicar la DWT
coeffs = pywt.wavedec(data, wavelet, level=level)

# Coeficientes de aproximación (cA) y detalle (cD) para cada nivel de descomposición
cA = coeffs[0]  # Coeficiente de aproximación en el último nivel
cD = coeffs[1:]  # Coeficientes de detalle en cada nivel

# Puedes usar cA y cD para análisis adicionales o reconstruir la señal original si es necesario
```
