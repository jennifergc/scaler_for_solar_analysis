## Scaler dataframe for Auger Open Data 
### Periodicity analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from augerscalerpy.scaler_clear import plotscaler 
```

```python
### PROMEDIO DE VENTANA MÓVIL 
### ____________________________UN DATO CADA MEDIA__________ TODOS LOS AÑOS (MEJORAR)

scaler3600s = plotscaler('scaler_ALL_hourly_11abril.csv','scaler_3600s')
```

```python

# Calcular promedio móvil cada X intervalo de tiempo

#window_size = "3600S"
#rolling_mean = scaler3600s['ScalerCorr'].rolling(window=window_size).mean()
# Crear el nuevo DataFrame con el promedio móvil
#promedio_24 = pd.DataFrame({'Original': data['ScalerCorr'], 'Rolling Mean': rolling_mean})

promedio_24 = scaler3600s['ScalerCorr'].rolling(24).mean().iloc[::24]

# Graficar curva ajustada al promedio de los datos cada 1000 datos
# Crear figura y especificar tamaño
fig = plt.figure(figsize=(20,10))
plt.plot(promedio_24.index, promedio_24.values,marker='.', linestyle='', label='Ajuste simple')
plt.xlabel('Tiempo')
plt.ylabel('ScalerCorr')
plt.title('Scaler Observatorio Pierre Auger')
plt.legend()
plt.show()
```
