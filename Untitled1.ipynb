```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly.graph_objects as go
# Importar las bibliotecas necesarias
import circlify
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("netflix.csv") ####### EL ARCHIVO DE SCALER PROCESADOS QUE ESSSS
#COMPROBANDO SI HAY VALORES NULOS
df
```

```python

# Asumiendo que df es tu DataFrame y tiene las columnas 'language', 'genre' y 'imdb_score'
df_grouped = df.groupby(['language', 'genre']).agg({'imdb_score': 'mean'}).reset_index()
```

```python
df_grouped
```

```python
# Crear una columna de padres para cada idioma y género
df_grouped['parent'] = df_grouped['language']
```

```python
df_grouped
```

```python

# Crear una columna de ids para cada género
df_grouped['id'] = df_grouped['language'] + '-' + df_grouped['genre']
df_grouped
```

```python
# Crear un DataFrame para cada idioma
df_languages = pd.DataFrame(df_grouped['language'].unique(), columns=['id'])

# Añadir las columnas 'parent' e 'imdb_score'
df_languages['parent'] = ''
df_languages['imdb_score'] = 0

# Concatenar df_grouped y df_languages
df_grouped = pd.concat([df_grouped, df_languages], ignore_index=True)
```

```python
# Importar las bibliotecas necesarias
import circlify
import matplotlib.pyplot as plt

# Agrupar los datos y calcular el número de películas para cada combinación de idioma y género
data = df.groupby(['language', 'genre']).size().reset_index(name='count')

# Crear una lista de diccionarios, donde cada diccionario representa un círculo en el diagrama
circles_data = []
for _, row in data.iterrows():
    circles_data.append({
        'datum': row['count'], 
        'id': f"{row['language']}-{row['genre']}", 
        'parent': row['language']
    })

# Calcular las posiciones de los círculos
circles = circlify.circlify(
    circles_data, 
    show_enclosure=False, 
    target_enclosure=circlify.Circle(x=0, y=0, r=1)
)

# Crear la figura y el eje
fig, ax = plt.subplots(figsize=(10, 10))

# Dibujar cada círculo
for circle in circles:
    x, y, r = circle.circle
    label = circle.ex['id']
    ax.add_patch(plt.Circle((x, y), r, alpha=0.5, linewidth=2))
    plt.annotate(label, (x, y), va='center', ha='center')

# Configurar los límites del eje y el título
ax.set_title('Diagrama de Árbol Circular de Películas por Idioma y Género')
ax.axis('off')
plt.xlim(-1.1, 1.1)
plt.ylim(-1.1, 1.1)

# Mostrar la gráfica
plt.show()
```
